```python
import pandas as pd

# 1. Make a Pandas DataFrame with a two-dimensional Python list
data = [
    [1, "Abhinav", 20],
    [2, "Prateek", 21],
    [3, "Harshit", 22]
]

df = pd.DataFrame(data)
print("Two-dimensional Python List DataFrame:")
print(df)


# 2. Create DataFrame from Python dict
data = {
    "name": ["Abhinav"],
    "surname": ["Sharma"]
}

df = pd.DataFrame(data)
print("\nDataFrame from Python Dict:")
print(df)


# 3. Create Pandas dataframe using list of lists
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

df = pd.DataFrame(data)
print("\nDataFrame using List of Lists:")
print(df)


# 4. Create a Pandas dataframe using list of tuples
data = [
    ("Abhinav", "Sharma", 20),
    ("Harshit", "Sharma", 25),
    ("Prateek", "Sharma", 28)
]

df = pd.DataFrame(data)
print("\nDataFrame using List of Tuples:")
print(df)


# 5. Create a Pandas DataFrame from List of Dicts
data = [
    {"Name": "Abhinav", "Age": 20},
    {"Name": "Harshit", "Age": 21},
    {"Name": "Prateek", "Age": 22}
]

df = pd.DataFrame(data)
print("\nDataFrame from List of Dicts:")
print(df)
```
