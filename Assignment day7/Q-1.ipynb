```python
import pandas as pd

# 1. Create a Pandas Series from Dictionary
data_dict = {
    "name": "Abhinav",
    "surname": "Sharma"
}

series_dict = pd.Series(data_dict)

print("Series from Dictionary:")
print(series_dict)


# 2. Create a Pandas Series from List
data_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

series_list = pd.Series(data_list)

print("\nSeries from List:")
print(series_list)


# 3. Access the elements of a Series
print("\nAccessing Elements:")
print(series_list[5])
print(series_list[1])
print(series_list[2])
print(series_list[8])
```
