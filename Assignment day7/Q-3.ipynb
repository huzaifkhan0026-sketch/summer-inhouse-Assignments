```python
import pandas as pd

# 1. Different ways to iterate over rows in Pandas DataFrame
data = {
    "Name": ["Abhinav", "Harshit", "Prateek"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

print("Iterating over rows:")
for index, row in df.iterrows():
    print(index, row["Name"], row["Age"])


# 2. Selecting rows in pandas DataFrame based on conditions
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [4, 5, 6, 7, 8],
    'C': [9, 0, 11, 12, 13]
})

result = df[df["C"] != 0]

print("\nRows based on condition:")
print(result)


# 3. Select any row from a DataFrame using iloc[]
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}

df = pd.DataFrame(data)

print("\nRow using iloc[]:")
print(df.iloc[0:1, :])


# 4. Limited rows selection with given column
data = {
    "Name": ["Abhinav", "Rahul", "Aman"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

print("\nLimited rows with given column:")
print(df.loc[0:1, ["Name"]])


# 5. Drop rows from the dataframe based on condition
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [4, 5, 6, 7, 8],
    'C': [9, 0, 11, 12, 13]
})

rows = df[df["C"] == 0].index
result = df.drop(rows)

print("\nDataFrame after dropping rows:")
print(result)


# 6. Insert row at given position in Pandas DataFrame
data = {
    "Name": ["Abhinav", "Rahul"],
    "Age": [20, 21]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    "Name": ["Aman"],
    "Age": [22]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print("\nDataFrame after inserting row:")
print(df)


# 7. Create a list from rows in Pandas DataFrame
data = {
    "Name": ["Abhinav", "Rahul", "Aman"],
    "Age": [20, 21, 22]
}

df = pd.DataFrame(data)

result = df.values.tolist()

print("\nList from DataFrame rows:")
print(result)
```
