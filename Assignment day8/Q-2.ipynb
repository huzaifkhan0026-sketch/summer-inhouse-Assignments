```python
import pandas as pd

# Question: Create two DataFrames, df1 and df2, with a common column (e.g., 'ID')

df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["B1", "B2", "B3", "B4"],
    "Subject_ID": ["sub1", "sub2", "sub3", "sub4"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5, 7],
    "Name2": ["A1", "A2", "A3", "A4", "A5", "A77"],
    "Subjects_ID": ["Subj8", "Subj3", "Subj9", "Subj99", "Subject33", "Subj323"]
})

print("DF1")
print(df1)

print("\nDF2")
print(df2)


# Question: Perform an inner merge on this common column and display the resulting DataFrame

inner_merge = pd.merge(df1, df2, on="ID", how="inner")

print("\nInner Merge Result")
print(inner_merge)


# Question: Perform a left join of df1 and df2 on the 'ID' column

left_join = pd.merge(df1, df2, on="ID", how="left")

print("\nLeft Join Result")
print(left_join)

# Missing values are handled as NaN when matching data is not found


# Question: Perform a right join using pd.merge() on a common column

right_join = pd.merge(df1, df2, on="ID", how="right")

print("\nRight Join Result")
print(right_join)


# Question: Perform a join using df.join() based on the index

left = df1.set_index("ID")
right = df2.set_index("ID")

index_join = left.join(
    right,
    lsuffix="_left",
    rsuffix="_right"
)

print("\nIndex Based Join Result")
print(index_join)


# Question: Compare the results

print("\nRight Join keeps all rows from df2.")
print("Index-Based Join combines data using index values.")


# Question: Merging with Multiple Keys

df3 = pd.DataFrame({
    "ID": [1, 2, 3],
    "City": ["Delhi", "Mumbai", "Pune"],
    "Marks": [80, 85, 90]
})

df4 = pd.DataFrame({
    "ID": [1, 2, 3],
    "City": ["Delhi", "Mumbai", "Pune"],
    "Grade": ["A", "B", "A+"]
})

multiple_key_merge = pd.merge(df3, df4, on=["ID", "City"])

print("\nMerge with Multiple Keys")
print(multiple_key_merge)
```
