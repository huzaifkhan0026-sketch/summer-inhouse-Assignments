```python
import pandas as pd

# Question 3: Create three DataFrames

df1 = pd.DataFrame({
    "ID": [1, 2],
    "Name": ["A1", "A2"]
})

df2 = pd.DataFrame({
    "ID": [3, 4],
    "Name": ["A3", "A4"]
})

df3 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Subject": ["Math", "Science", "English", "Physics"]
})


# Question: Vertically concatenate two DataFrames using pd.concat()

concat_df = pd.concat([df1, df2], axis=0)

print("Concatenated DataFrame:")
print(concat_df)


# Question: Merge the resulting DataFrame with the third DataFrame on a common key

final_df = pd.merge(concat_df, df3, on="ID")

print("\nMerged DataFrame:")
print(final_df)


# Question: Understand join() vs merge()

left = df1.set_index("ID")
right = df3.set_index("ID")

join_result = left.join(right)

print("\nJoin Result:")
print(join_result)


merge_result = pd.merge(df1, df3, on="ID")

print("\nMerge Result:")
print(merge_result)


# Question: Explain the primary differences between df.join() and pd.merge()

print("\nPrimary Differences Between join() and merge()")
print("1. join() works mainly on index.")
print("2. merge() works mainly on common columns.")
print("3. join() is simpler for index-based joining.")
print("4. merge() is more flexible and supports multiple keys.")
```
