```python
import pandas as pd

# Import CSV File
df = pd.read_csv("tips (1).csv")

print("Original Data:")
print(df)

# Data Cleaning
print("\nMissing Values:")
print(df.isnull().sum())

# Replace missing values
df = df.fillna(0)

print("\nCleaned Data:")
print(df)

# Data Analysis
print("\nSummary Statistics:")
print(df.describe())

# Average Tip
print("\nAverage Tip:")
print(df['tip'].mean())

# Maximum Bill
print("\nMaximum Total Bill:")
print(df['total_bill'].max())

# Minimum Bill
print("\nMinimum Total Bill:")
print(df['total_bill'].min())

# Total Tips Collected
print("\nTotal Tips:")
print(df['tip'].sum())

# Average Tip by Gender
print("\nAverage Tip by Gender:")
print(df.groupby('sex')['tip'].mean())
```
