```python
import numpy as np

# Convert 1D array to 2D
arrr = np.array([10,20,30,40,50,60])
res = arrr.reshape(2,3)
print(arrr)
print(res)

# Array Attributes
print(res.shape)
print(res.dtype)
print(res.itemsize)

# Create a 3×3 NumPy array of all 9
arr3 = np.full((3,3), 9)
print(arr3)

# OR

arr3 = np.arange(9)
print(arr3)
res = arr3.reshape((3,3))
print(res)

# Create a 1D array of 10 evenly spaced values between 25 and 125
bwt = np.linspace(25,125,10)
print(bwt)

# Convert a Python list into a NumPy array
list = [1,2,3,4,5,6,7,8]
print(list)

resv = np.array(list)
print(resv)

# Reverse a 1D NumPy array
fro = np.array([1,2,3,4,5,6,7,8])
print(fro)

rev = np.flip(fro)
print(rev)

# Create a 4×4×3 array and extract value at second set, first row and last column
res = np.arange(48)
print(res)

res1 = res.reshape(4,4,3)
print(res1)

res2 = res1[1,0,-1]
print(res2)

# Create a 4×4 and Extract Odd Rows and Even Columns
r = np.arange(16)
print(r)

res = r.reshape(4,4)
print(res)

res2 = res[::2,1::2]
print(res2)

# Slice first two rows and first two columns of second set from a 4×4×3 array
res = np.arange(48)
print(res)

res1 = res.reshape(4,4,3)
print(res1)

res2 = res1[1,0:2,0:2]
print(res2)

# Replace all odd numbers with -1 using for loop
array = np.array([[23,56,78,93],
                  [71,82,13,14]])

for i in range(array.shape[0]):
    for j in range(array.shape[1]):
        if array[i][j] % 2 != 0:
            array[i][j] = -1

print(array)

# Get indices of non-zero elements
arra = np.array([1,0,2,0,3,0,4])

for i in range(len(arra)):
    if arra[i] != 0:
        print(i)

# Arithmetic operations on two NumPy arrays
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9,10])

print(arr1)
print(arr2)

print(arr1 + arr2)
print(arr1 * arr2)

# Dot Product
array1 = np.array([15,20,25])
array2 = np.array([10,40,37])

print(array1, array2)

res = np.dot(array1, array2)
print(res)
```
