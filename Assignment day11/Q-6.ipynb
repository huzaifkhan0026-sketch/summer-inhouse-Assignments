```python
import numpy as np

arr1 = np.array([
    [[41, 42], [43, 44]],
    [[45, 46], [47, 48]]
])

arr2 = np.array([
    [[29, 30], [31, 32]],
    [[33, 34], [35, 36]]
])

avg = (arr1 + arr2) / 2

print("Average of NumPy arrays:\n", avg)

print("\nMean of avg\n", np.mean(avg))

print("\nMedian of avg\n", np.median(avg))

values, counts = np.unique(avg, return_counts=True)
mode = values[np.argmax(counts)]

print("\nMode of avg\n", mode)
```

```python
import numpy as np
import pandas as pd

arr1 = np.array([
    [[41, 42], [43, 44]],
    [[45, 46], [47, 48]]
])

arr2 = np.array([
    [[29, 30], [31, 32]],
    [[33, 34], [35, 36]]
])

avg = (arr1 + arr2) / 2

mode = pd.Series(avg.reshape(-1)).mode()

print("Average of NumPy arrays:\n", avg)
print("\nMean =", np.mean(avg))
print("\nMedian =", np.median(avg))
print("\nMode =\n", mode)
```
