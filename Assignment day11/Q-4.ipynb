```python
import numpy as np

rr = np.array([1212, 24, 48, 45, 6])

print("Array:", rr)
print("Maximum Value:", np.max(rr))
print("Minimum Value:", np.min(rr))

for i in rr:
    print(i)
```

```python
import numpy as np

arrg = np.array([[10, 20, 30],
                 [40, 50, 60]])

print("Element at row 1, column 2:", arrg[1, 2])

total = 0
rows, cols = arrg.shape

for row in arrg:
    for value in row:
        total += value

print("Rows:", rows)
print("Columns:", cols)

print("Array:")
print(arrg)

print("Sum =", total)
```
