```python
# 8.Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

def prime(num):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                print("Not Prime")
                break
        else:
            print("Prime")
    else:
        print("Not Prime")

prime(4)
```
