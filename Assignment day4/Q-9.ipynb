```python
#Write a Python function that accepts a string and counts 
#the number of upper and lower case letters. 

def count_case(str):
    toupper = 0
    tolower = 0

    for i in str:
        if i.isupper():
            toupper += 1
        elif i.islower():
            tolower += 1

    print("Upper Case Letters:", toupper)
    print("Lower Case Letters:", tolower)

count_case("Hello Python")
```
