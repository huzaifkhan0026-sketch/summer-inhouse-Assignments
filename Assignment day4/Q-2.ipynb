```python
# 2. Write a Python function that takes a list and returns a new list with distinct elements from the first list.
def distinct_list(lst):
    new_list = []

    for i in lst:
        if i not in new_list:
            new_list.append(i)

    return new_list

print(distinct_list([1, 2, 3, 2, 4, 1, 5]))
```
