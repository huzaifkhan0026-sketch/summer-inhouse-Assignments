```python
import pandas as pd
import numpy as np

a = np.array([
    [11, 21, np.nan],
    [32, np.nan, 45],
    [np.nan, 58, 67]
])

df = pd.DataFrame(a)

print("Original DataFrame:")
print(df)

# Replace NaN with column mean
df.fillna(df.mean(), inplace=True)

print("\nAfter Replacing NaN with Column Average:")
print(df)
```
