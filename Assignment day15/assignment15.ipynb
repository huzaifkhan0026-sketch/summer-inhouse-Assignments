```python
t = (28, 29, 50, 10, 35)
total = 0
max = t[0]
min = t[0]
for i in t:
    total = i + total

    if i > max:
        max = i

    if i < min:
        min = i

print("Total =", total)
print("Highest =", max)
print("Lowest =", min)
```
