```python
#3
import numpy as np

arr1 = np.array([[6, -8, 73, -110],
                 [np.nan, -8, 0, 94]])

print(arr1)

mean = np.nanmean(arr1, axis=0)

index = np.where(np.isnan(arr1))

arr1[index] = np.take(mean, index[1])

print(arr1)
```

```python
#5
import numpy as np

arr1 = np.array([[1,2],
                 [3,4],
                 [5,6],
                 [7,8]])

arr2 = np.array([[11,22],
                 [33,44],
                 [55,66],
                 [77,88]])

avg = (arr1 + arr2) / 2

print("Average of the array is:\n", avg)

print("Mean of the array is =", np.mean(avg))

print("Median of the array is =", np.median(avg))

values, counts = np.unique(avg, return_counts=True)
mode = values[np.argmax(counts)]

print("Mode of the array is =", mode)
```

```python
#6
import numpy as np

a = np.array([[1, -2, 3],
              [-1, 3, -1],
              [2, -5, 5]])

b = np.array([9, -6, 17])

# Using linalg.solve()
res = np.linalg.solve(a, b)
print(res)

# Using Inverse Matrix Method
a_inv = np.linalg.inv(a)
print(a_inv)

res = np.dot(a_inv, b)
print(res)
```
